```python
import lime
import lime.lime_tabular
```

```python
method2allimport = {}

method = 'gatv2_exp_layer_1'
def get_index(name, lst):
    for i, l in enumerate(lst):
        if name == l:
            return i
def get_contain_attris(attr_list, target_list):
    for attr in attr_list:
        for tar in target_list:
            if attr in tar:
                return tar
dataset2example_idx = {
    'ohio': 150 + 258,
    'arises': 0 + 244, 
    'shanghai_t1dm': 0 + 113,
    'shanghai_t2dm' : 0 + 74,  
}
dataset2example_pid = {
    'ohio': 591, 
    'arises' : 1005,
    'shanghai_t1dm' : 1001,
    'shanghai_t2dm' : 2030,

}
for dataset in ['ohio', 'arises', 'shanghai_t1dm', 'shanghai_t2dm']:
    # for dataset in ['ohio', 'arises']:
    print(dataset)
    seed = 1
    from gat_mem_org_log.train import *
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False
    os.environ['PYTHONHASHSEED'] = str(seed)

    version = 'nips_23'

    CONF = select_conf(dataset=dataset, version=version, seed=seed, method=method, hidden_dim=mthd_dt2hidden_dim[(method, dataset)])

    log = Log(CONF['dataset'] + '_' + CONF['model_name'] + '_' + CONF['comments'], CONF)
    data_proc = DataProcessor(CONF)



    pid_list = list(data_proc.train_pid2data_npy.keys())
    update = Update(data_proc, CONF)
    attris_name = ['glucose_level'] + CONF['attri_list'] + CONF['time_attri_list']

    pid2model = {}
    pid_lime_values = []
    pid_example_shape_values = np.zeros((1, CONF['seq_len'], len(attris_name)), dtype=np.float32)

    for pid in pid_list:
        pid2model[pid] = get_model(CONF)
        pid2model[pid] = log.load_models(pid2model[pid], CONF, pid)

        pid2model[pid].to(CONF['device'])


        G, attris, time_attris, y, target_time = data_proc.get_all_train(CONF['device'], CONF['seq_len'],  CONF['attri_list'], CONF['time_attri_list'], pid )

        X_train = torch.concatenate([G.unsqueeze(dim=-1), attris, time_attris], dim=-1)
        # X_train_idx = np.random.choice(X_train.shape[0], 100, replace=False)

        explainer = lime.lime_tabular.RecurrentTabularExplainer(X_train.detach().cpu().numpy(),training_labels=y, 
                                                            feature_names=attris_name, verbose=True, mode='regression')
        B = 100

        G, attris, time_attris, y, target_time, seq_st_ed = data_proc.get_test(CONF['device'], CONF['seq_len'], CONF['attri_list'], CONF['time_attri_list'], pid)
        X_test = torch.cat([G.unsqueeze(dim=-1), attris, time_attris], dim=-1).detach().cpu().numpy()
        
        def predict(X_test):
            X_test = torch.tensor(X_test, dtype=torch.float32, device=CONF['device'])
            pred = pid2model[pid](X_test)
            pred = pred.detach().cpu().numpy()
            return pred
        try:
            X_train_idx = np.random.choice(X_train.shape[0], B, replace=False)
        except:
            continue
        all_method_all_attn_probs = np.zeros((B, X_test.shape[1], X_test.shape[2]), dtype=np.float32) # B, T, C
        X_train = X_train.detach().cpu().numpy()
        for b, j in enumerate(X_train_idx):
            exp = explainer.explain_instance(X_train[j], predict, num_features=48 * 16)
            exp_list = exp.as_list()
            for item in exp_list:
                value = item[1]
                sub_item = get_contain_attris(attris_name, item[0].split(' '))

                t = int(sub_item.split('-')[-1])
                c_name = sub_item.split('-')[0][:-2]
                c = get_index(c_name, attris_name)
                all_method_all_attn_probs[b, t, c] = value
        
        j = dataset2example_idx[dataset]
        sample_pid = dataset2example_pid[dataset]
        if pid == sample_pid:
            exp = explainer.explain_instance(X_test[j], predict, num_features=48 * 16)
            exp_list = exp.as_list()
            for item in exp_list:
                value = item[1]
                sub_item = get_contain_attris(attris_name, item[0].split(' '))

                t = int(sub_item.split('-')[-1])
                c_name = sub_item.split('-')[0][:-2]
                c = get_index(c_name, attris_name)
                pid_example_shape_values[0, t, c] = value
        pid_lime_values.append(all_method_all_attn_probs)
 
    save_path = os.path.join(CONF['log_path'], 'LIME')
    if not os.path.exists(save_path):
        os.makedirs(save_path)
    lime_values = np.concatenate(pid_lime_values, axis=0)

    np.save(save_path+'/example_importance_values.npy', pid_example_shape_values)
    np.save(save_path+'/all_importance_values.npy', lime_values)
```
